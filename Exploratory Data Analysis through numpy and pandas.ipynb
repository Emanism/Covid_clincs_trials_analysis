```python
# 5 Questions Using Group By
```

```python
#1.How many clinical trials are there for each condition?
```

```python
trials_per_condition = df.groupby('Conditions')['NCT Number'].count()
print(trials_per_condition)
```

```python
#2.How many trials are there for each status?
```

```python
trials_per_status = df.groupby('Status')['NCT Number'].count()
print(trials_per_status)
```

```python
df.columns
```

```python
#3.How many different sponsors are there for each condition?
```

```python
sponsors_per_condition = df.groupby('Conditions')['Sponsor/Collaborators'].nunique()
print(sponsors_per_condition)
```

```python
#4.What is the count of clinical trials in each country?
```

```python
df['Country'] = df['Locations'].str.split(', ').str[-1]
trials_per_country = df.groupby('Country')['NCT Number'].count()
print(trials_per_country)
```

```python
#5.What is the earliest start date for clinical trials by condition?
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')
earliest_start_by_condition = df.groupby('Conditions')['Start Date'].min()
print(earliest_start_by_condition)
```

```python
#Make 5 questions and solve them by using conditional statements
```

```python
#1.How many clinical trials are 'Recruiting' and have 'No Results Available'?
```

```python
recruiting_no_results = df[(df['Status'] == 'Recruiting') & (df['Study Results'] == 'No Results Available')].shape[0]
print(recruiting_no_results, 'have no results availabe')
```

```python
#2.How many clinical trials involve 'Drug' interventions as compared to other types of interventions?
```

```python
drug_interventions = df[df['Interventions'].str.contains('Drug', na=False)].shape[0]
other_interventions = df[~df['Interventions'].str.contains('Drug', na=False)].shape[0]
print('drug_interventions :', drug_interventions )
print('other_interventions: ', other_interventions )
```

```python
#3.How many clinical trials started after January 1, 2020?
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')
trials_post_2020 = df[df['Start Date'] > pd.Timestamp('2020-01-01')].shape[0]
print(trials_post_2020, 'trials started after jan 1')
```

```python
#4.How many clinical trials are there for each condition that has 'COVID' in their title?
```

```python
covid_in_title = df[df['Title'].str.contains('COVID', case=False, na=False)].shape[0]
print(covid_in_title, 'clinical trials were there for covid')
```

```python
#5.How many clinical trials have their primary completion date in the future (from April 2023)?
```

```python
from datetime import datetime
future_date = datetime(2023, 4, 1)
df['Primary Completion Date'] = pd.to_datetime(df['Primary Completion Date'], errors='coerce')
future_completion = df[df['Primary Completion Date'] > future_date].shape[0]
print(future_completion,'clinics have completion date in the future')
```
